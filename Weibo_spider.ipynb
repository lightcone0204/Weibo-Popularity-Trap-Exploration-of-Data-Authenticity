# 真假流量所占的比例各有多少？

```python
data_fake = data[((data['user.follow_count']<=5)|(data['user.followers_count']<=5))&
                 (data['user.description']=='')&
                 (data['comments_count']==0)&
                (data['attitudes_count']==0)&
                (data['reposts_count']==0)&
                (data['user.mbrank']==0)]
data_fake.sample(5)
```

```python
data_fake.shape
```

```python
# 昵称里包含“用户”的，基本上可以断定是假粉丝
data_fake2_index = data[(data['user.follow_count']>5)&
                        (data['user.followers_count']>5)&
                        (data['user.screen_name'].str.contains('用户'))].index
```

```python
data_fake = pd.concat([data_fake, data.iloc[data_fake2_index]])
data_true = data.drop(data_fake.index)
data_true= data_true.drop_duplicates(subset=['user.id'],keep='first')
data_fake=data.drop(data_true.index)
```

```python
data_true["F_or_T"]="T"
data_fake["F_or_T"]="F"
```

```python
data2 = data_true.append(data_fake)
data2.to_csv('{}识别数据.csv'.format(name), index=False,encoding='utf_8_sig')
```

```python
print('真粉丝转发数占总转发数的{}%'.format(np.round(data_true.shape[0]/data.shape[0]*100, 2)))
print('假粉丝转发数占总转发数的{}%'.format(np.round(data_fake.shape[0]/data.shape[0]*100, 2)))
```

# 真流量粉的粉丝画像

```python
# 绘制蔡徐坤真粉丝的简介词云图
import jieba
from collections import Counter
from pyecharts.charts import WordCloud 

jieba.add_word('{}'.format(name))

swords = [x.strip() for x in open ('stopwords.txt',encoding='utf-8')]
```

```python
def plot_word_cloud(data, swords, columns):
    text = ''.join(data[columns])
    words = list(jieba.cut(text))
    ex_sw_words = []
    for word in words:
        if len(word)>1 and (word not in swords):
            ex_sw_words.append(word)
    c = Counter()
    c = Counter(ex_sw_words)
    wc_data = pd.DataFrame({'word':list(c.keys()), 'counts':list(c.values())}).sort_values(by='counts', ascending=False).head(100)
    a = pd.DataFrame({'word':wc_data['word'],'counts':wc_data['counts']})
    words=[(i,j) for i,j in zip(wc_data['word'], wc_data['counts'])]
    wordcloud = WordCloud()
    wordcloud.add("",words,word_size_range=[20, 100])
    return wordcloud,a
```

```python
a,b=plot_word_cloud(data=data_true, swords=swords, columns='user.description')
a.render('{}粉丝用户描述词云.html'.format(name))
b.to_csv('{}用户描述词云数据.csv'.format(name), index=False,encoding='utf_8_sig')
```

```python
a,b=plot_word_cloud(data=data_true, swords=swords, columns='raw_text')
a.render('{}粉丝转发信息词云.html'.format(name))
b.to_csv('{}用户转发词云数据.csv'.format(name), index=False,encoding='utf_8_sig')
```
